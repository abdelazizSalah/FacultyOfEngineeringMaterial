```python
from commonfunctions import *
import math
from skimage import filters
from numpy import sort
%matplotlib inline
%load_ext autoreload
%autoreload 2

#reading an image 
img = io.imread('./pout.tif')
io.imshow(img)
io.show()

# applying noise on the img
noisyImg = random_noise(image = img, mode= 's&p',amount =0.05 )
io.imshow(noisyImg)
io.show()

```

```python
'''
(1) Your own median filter algorithm
'''
import copy

edgex = 1
edgey = 1 

# y is rows 
# x is columns
img2 = copy.deepcopy(noisyImg) # we could use np.copy instead but I found this new lib so I tried to use it. 


# printing the noisy img
io.imshow(img2)
io.show()

# for y from egdey to the whole height
for col in range (edgey, noisyImg.shape[0] - edgey + 1): 
    # for x from edgex to the whole width
    for row in range (edgex, noisyImg.shape[1] - edgex + 1 ):
        colorArr = np.array(noisyImg[col: col+3, row : row  + 3])
        colorArrFlatten = colorArr.flatten()
        colorArrFlatten = sort(colorArrFlatten)
        # output pixed should be in the middle
        img2[col][row] = np.median(colorArrFlatten)


# printing the img after applying the algo.
io.imshow(img2)
io.show()
        
```
