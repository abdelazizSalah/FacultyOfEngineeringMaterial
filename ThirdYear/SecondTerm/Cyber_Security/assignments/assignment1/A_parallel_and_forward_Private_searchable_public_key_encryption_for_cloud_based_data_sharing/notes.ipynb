* The problem which the paper try to solve is:

* what is the forward security:
  * Forward security, also known as forward secrecy, is a property of cryptographic protocols that ensures that the compromise of long-term secret keys does not allow an attacker to decrypt past communication sessions that were previously encrypted using those keys.

  *   In other words, forward security means that even if an attacker gains access to a secret key used to encrypt past communication, they will not be able to use that key to decrypt past communications.

  *   Forward security is important because it limits the amount of information an attacker can gain from a compromised key. If forward security is not employed, an attacker who compromises a long-term secret key can decrypt all communication that was encrypted using that key, potentially compromising a large amount of sensitive information.

  *   Forward security is often achieved through the use of ephemeral keys, which are temporary keys that are generated for each individual communication session. Ephemeral keys are never used again after the session is completed, so even if they are compromised, they cannot be used to decrypt past sessions.

### Bilinear mapping e
* Bilinear pairing refers to a mathematical operation that takes two elements from two different groups and maps them to an element in a third group. This operation is usually denoted by the symbol "e" and is commonly used in cryptography and number theory.

## Difference between public channel and private channel 
A public channel and a secured channel are two different types of communication channels used in computer networks and cryptography.

A public channel is a communication channel that is accessible to anyone on the network. Information sent over a public channel is not encrypted or protected, and can be intercepted and read by anyone with access to the channel. Examples of public channels include social media platforms, email services, and public Wi-Fi networks.

In contrast, a secured channel is a communication channel that is encrypted and protected to prevent unauthorized access and ensure the confidentiality and integrity of the information being transmitted. A secured channel typically requires some form of authentication or verification to ensure that only authorized parties can access the channel. Examples of secured channels include VPNs (Virtual Private Networks), encrypted messaging apps, and secure file transfer protocols.

The main difference between a public channel and a secured channel is the level of protection and privacy they offer. Public channels offer no encryption or protection, while secured channels provide strong encryption and authentication mechanisms to ensure that the information being transmitted remains confidential and secure.
